```python
# Reads in names, indexes and codes from worksheet downloaded from BMF
df_codes = pd.read_excel(path+'AcoesIndices.SetDez.2019.xlsx', header = 3, index_col = 0)
```

```python
# Size of data frame
df_codes.shape
```

```python
df_codes.head(3)
```

```python
# Use bfill() function to populate missing values na values in the dataframe across rows.
df_codes['codigo']=df_codes.bfill(axis=1).iloc[:, 1]
```

```python
df_codes.head(3)
```

```python
df_codes.tail()
```

```python
# Reorder data
neworder = [ 'codigo', 'Tipo', 'BDRX', 'IBOV', 'IBRA', 'IBXL', 'IBXX', 'ICO2', 'ICON', 'IDIV',
       'IEEX', 'IFIX', 'IFNC', 'IGCT', 'IGCX', 'IGNM', 'IMAT', 'IMOB', 'INDX',
       'ISEE', 'ITAG', 'IVBX', 'MLCX', 'SMLL', 'UTIL' ]

df_codes=df_codes.reindex(columns=neworder)
```

```python
df_codes.head()
```

```python
# Drop last row
df_codes = df_codes.drop(df_codes.tail(1).index)
df_codes.shape
```

```python
df_codes.tail()
```

```python
# Save data to csv files - changed separator to ';'
df_codes.to_csv(path+'codigos_acoes_bovespa.csv', sep = ';', columns = ['codigo'], encoding = 'utf-8')
```

## Open updated version of codes spreadsheet

```python
# Reads in names, indexes and codes from worksheet downloaded from BMF
df_codes2 = pd.read_excel(path+'AcoesIndices.MaiAgo.2020.xlsx', header = 3, index_col = 0)
```

```python
df_codes2.shape
```

```python
df_codes2['codigo']=df_codes2.bfill(axis=1).iloc[:, 1]

# Reorder data
neworder = [ 'codigo', 'Tipo', 'BDRX', 'IBOV', 'IBRA', 'IBXL', 'IBXX', 'ICO2', 'ICON', 'IDIV',
       'IEEX', 'IFIX', 'IFNC', 'IGCT', 'IGCX', 'IGNM', 'IMAT', 'IMOB', 'INDX',
       'ISEE', 'ITAG', 'IVBX', 'MLCX', 'SMLL', 'UTIL' ]

df_codes2=df_codes2.reindex(columns=neworder)
```

```python
df_codes2.tail()
```

```python
# Drop last row
df_codes2 = df_codes2.drop(df_codes2.tail(1).index)
df_codes2.shape
```

```python
# Save data to csv files - changed separator to ';'
df_codes2.to_csv(path+'codigos_acoes_bovespa_2020T2.csv', sep = ';', columns = ['codigo'], encoding = 'utf-8')
```

```python
## Concatenate rows of dataframes with codes
df_all = pd.concat([df_codes,df_codes2], axis = 'rows')
```
