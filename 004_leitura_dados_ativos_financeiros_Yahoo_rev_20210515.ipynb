# Usando o Yahoo Finance

## Parte 1 - ações (agora sem BDR) da lista da Bovespa

## Data de início: 01/01/2017

```python
# Alterando os símbolos para o Yahoo (adiciona .SA no final)
kk2 = [ i + '.SA' for i in kk]
```

```python
import datetime

start = datetime.datetime(2017,1,1)
end = datetime.datetime.now().date()

time_start = perf_counter()

df_adj_close = yf.download(kk2, start=start, end=end)['Adj Close']#, 'Close']]
df_close = yf.download(kk2, start=start, end=end)['Close']

time_end = perf_counter()
elapsed_time = time_end - time_start
print(' ')
print(' **** o programa rodou em ', round(elapsed_time,2) , ' segundos *****')
```
